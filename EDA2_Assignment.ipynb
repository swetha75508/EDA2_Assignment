# Feature Engineering

```python
# create age group feature
df['age_group'] = pd.cut(df['age'], bins=[18, 30, 45, 60, 90], labels=['youngsters(18-30)', 'adult(31-45)', 'middle age(46-60)', 'senior(61+)'])

# create work hours category feature
df['work_hours_category'] = pd.cut(df['hours_per_week'], bins=[20,40,60,70], labels= ['part-time', 'full-time', 'overtime'])

# check the new created features
print(df[['age_group','work_hours_category']])
```

# Feature Selection
Isolation Forest Algorithm

```python
# import isolation forest
from sklearn.ensemble import IsolationForest

clf = IsolationForest(random_state=42, contamination = 0.05)
# contamination = 0.05 -> specifies the proportion of outliers in the data
```

```python
# encode the entire actual dataset for model performance
df_encoded = pd.get_dummies(df)
df_encoded
```

```python
# train the model
outliers= clf.fit(df_encoded)
```

```python
# test the model
y_predict_outliers = clf.predict(df_encoded)
y_predict_outliers
```

```python
df_encoded
```

```python
df_outliers = clf.fit_predict(df.select_dtypes(include = ['int64','float64']))

df_no_outliers = df[df_outliers !=-1]
df_no_outliers
```

```python
print(f"Original shape: {df.shape}")
print(f"Cleaned shape: {df_no_outliers.shape}")
```
